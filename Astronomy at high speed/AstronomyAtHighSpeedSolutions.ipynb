## Exercise A-1

From our viewpoint, the circular orbit appears as shown below:

![Our view of circular binary orbit](A11.png "Our view of a circular binary orbit")

The point $B$ marks star 1 which is regarded as stationary here. I am also assuming that the plane of the sky is located at the same distance from us as star 1 so that the dashed line $NN'$ marks the intersection between the orbital plane of the binary and the plane of the sky. This line is known as the "line of nodes" in binary star work. The distance from point $B$ to $C$ is the minimum distance which we wish to determine.

Now consider a side-on view along the direction of the line of nodes, which schematically looks as follows:

![Same orbit seen from along the line of nodes](A12.png "Same orbit seen from along the line of nodes")

By definition, the line of nodes is located within the orbital plane, thus the circular orbit is seen edge-on and  appears as a line of total length = 2 (remember all lengths are scaled in terms of the orbital separation). This line is indicated in the figure by the sloped line ending in short lines at either end. Since the line of nodes is is the orbital plane and the plane of the sky, it is perpendicular to both the orbital axis and the vector to Earth, and so these can therefore can be drawn in the plane of the diagram. The orbital inclination, $i$, is the angle between these two lines.

With the geometry defined, the solution is straightfoward. Consider the triangle marked $ABC$. The angle
$\angle ABC$ is equal to the orbital inclination $i$, while the length $BC$ equals the orbital separation, i.e. 1 in our units. Hence seen in projection as the line $AB$, we see a minimum separation $= \cos(i)$.

## Exercise A-2

We made the assumption of *total* eclipses (assumption 3). The critical case where this is just satisfied is illustrated below:

![Critical case where a binary just shows total eclipses](A21.png "Critical case of a binary which barely shows total eclipses")

Here I have indicated a new important point, the "contact" point, by $X$. We know the distance $BC = \cos(i)$, but clearly $CX = r_2$ and $BX = r_1$. Hence in this critical case $BC = CX - BX$ implies $\cos(i) = r_2 - r_1$. For any higher inclination, i.e. $i$ closer to $90^circ$, the cosine will get smaller, and thus the general condition for total eclipses given the assumptions is $$\cos(i) \le r_2 - r_1.$$

## Exercise A-3

The critical case for two stars barely eclipsing looks like so:


![Critical case where a binary just shows any eclipses](A31.png "Critical case of a binary which barely shows any eclipse")

The point $X$ still labels the point of contact of the two stars (but always remember they are at different distances from us so they are not genuinely touching each other). Now one can see that $CB = CX + BX$, so $\cos(i) = r_1 + r_2$. With the same argument about cosine getting smaller, the condition for *any* eclipse to be seen, total or partial, is therefore
$$ \cos(i) \le r_1 + r_2 .$$

## Exercise B-1

The diagram for Ex. A-3 is the situation that occurs at first or fourth contact, when the centres of the two stars are separated by a distance $d$ equal to the sum of their radii $r_1 + r_2$ on the plane of the sky. The diagram for Ex. A-2 is the situation at second or third contact when the centres of the two stars are separated by their difference $r_2 - r_1$. The question specifies the relative $x$, $y$ positions of the two stars, so we use Pythagoras' theorem to obtain the centre-to-centre distance, $d$, i.e. $d^2 = x^2 + y^2$. Therefore at fourth contact, setting $\phi = \phi_4$ and $d = r_1 + r_2$ we obtain
$$ \sin^2(i) \sin^2(2\pi\phi_4) + \cos^2(i) = (r_2+r_1)^2, $$
whereas at third contact $\phi = \phi_3$ and $d = r_2 - r_1$ and therefore
$$ \sin^2(i) \sin^2(2\pi\phi_3) + \cos^2(i) = (r_2-r_1)^2.$$

## Exercise B-2

The figures were made with $i = 87^\circ$, $r_1 = 0.15$ and $r_2 = 0.25$. Re-arranging the equation involving $\phi_3$ we have
$$ 
\begin{align}
\sin(2\pi\phi_3) &= \sqrt{ \frac{(r_2-r_1)^2 - \cos^2(i)}{\sin^2(i)}},\\
 &= \sqrt{ \frac{(0.25-0.15)^2 - \cos^2(87^\circ)}{\sin^2(87^\circ)}},\\
 &= 0.08532813.\\
 \end{align}
$$
This then leads to $\phi_3 = 0.013597$. A similar calculation gives $\phi_4 = 0.064993$.

## Exercise B-3

No, $\phi_3$ and $\phi_4$ alone are not enough because there are 3 unknowns, i.e. $r_1$, $r_2$ and $i$. Put another way, there are multiple set of $r_1$, $r_2$ and $i$ that give the same values of $\phi_3$ and $\phi_4$, so $\phi_3$ and $\phi_4$ alone do not fix $r_1$ and $r_2$ uniquely.

Since $\phi_1 = -\phi_4$ and $\phi_2 = - \phi_3$ by symmetry, they provide no extra information for the determination of $r_1$ and $r_2$.

## Exercise B-4

$f_1$ is the flux at primary eclipse when star 1 is entirely obscured, therefore we only see star 2 and so 
$f_1 = \pi r_2^2 s_2$. $f_2$ is the flux at secondary eclipse when we see all of star 1, but only some of star 2, since an area of $\pi r_1^2$ is hidden by star 1. Thus $f_2 = \pi (r_2^2 - r_1^2) s_2 + \pi r_1^2 s_1$. All three results can be summarised as:
$$
\begin{align}
f_0 &= \pi r_1^2 s_1 + \pi r_2^2 s_2,\\
f_1 &= \pi r_2^2 s_2,\\
f_2 &= \pi r_2^2 s_2 + \pi r_1^2 (s_1 - s_2).
\end{align}
$$

## Exercise B-5

Taking the difference between $f_0$ and $f_2$:
$$f_0 - f_2 = \pi r_1^2 s_2.$$
(This is the amount of flux from star 2 blocked by star 1.) Dividing this by $f_1$ eliminates $s_2$ to give
$$ \frac{f_0 - f_2}{f_1} = \frac{r_1^2}{r_2^2} . $$
The result asked for follows directly.
